```python
import random
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import style
%matplotlib inline


style.use('seaborn-white')

fig = plt.figure()
```

```python
def create_plots():
    xs = []
    ys = []
    
    for i in range(10):
        x = i
        y = random.randrange(10)
        xs.append(x)
        ys.append(y)
    return xs, ys
```

```python

ax1 = fig.add_subplot(2, 1, 1)
ax1.text(0.5, 0.5, str((2, 3, 1)),
           fontsize=18, ha='center')
```

```python
ax2 = fig.add_subplot(2, 1, 2)

x,y = create_plots()
ax1.plot(x,y)

x,y = create_plots()
ax2.plot(x,y)

# plt.show()
```

```python
fig = plt.figure()
ax1 = fig.add_axes([0.1, 0.5, 0.8, 0.4],
                   xticklabels=[], ylim=(-1.2, 1.2))
ax2 = fig.add_axes([0.1, 0.1, 0.8, 0.4],
                   ylim=(-1.2, 1.2))

x = np.linspace(0, 10)
ax1.plot(np.sin(x))
ax2.plot(np.cos(x));
```
